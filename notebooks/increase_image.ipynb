```
import cv2
import numpy as np
import pandas as pd
import os
import glob
```

```
csv = pd.read_csv('food/labels.csv')
csv
```

```
def save_img(path, src):
    dir = os.path.dirname(path)
    file_len = len(glob.glob(dir + '/*'))
    path = '{}/{:0=4}.jpg'.format(dir, file_len)
    cv2.imwrite(path, src)
    return path

def add_label(label, path, csv):
    new = pd.DataFrame([[label, path]], columns=['label', 'path'])
    csv = csv.append(new, ignore_index=True)
    return csv
```

```
new_csv = csv
for row in csv.itertuples():
    label = row[1]
    path = row[2]

    src = cv2.imread(path, 1)
    src = cv2.flip(src, 0)

    path = save_img(path, src)
    new_csv = add_label(label, path, new_csv)

new_csv.to_csv('food/labels.csv', index=None)

csv = pd.read_csv('food/labels.csv')
csv
```
