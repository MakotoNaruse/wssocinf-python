```
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels = []
```

```
food = 'tacorice'
files = glob.glob('food/{}/*'.format(food))
for f in files:
    image = plt.imread(f)
    plt.imshow(image)
    plt.show()
    labels.append('{},{}'.format(input(), f))
    plt.close()
```

```
np.savetxt('labels.csv', np.array(labels), fmt='%s', header='label,path')
```
